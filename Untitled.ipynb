```python
import numpy as np
import matplotlib.pylab as plt
```

```python
data1 = np.genfromtxt("Adobe1.txt",skip_header=2)
data2 = np.genfromtxt("Adobe2.txt",skip_header=2)
data3 = np.genfromtxt("Adobe3.txt",skip_header=2)
data4 = np.genfromtxt("32.txt",skip_header=2)
data5 = np.genfromtxt("17.txt",skip_header=2)
plt.figure()
plt.plot(data1[:,2],data1[:,3], '.-', color="dodgerblue")
plt.plot(data2[:,2],data2[:,3], '.-', color="navy")
plt.plot(data3[:,2],data3[:,3], '.-', color="blue")
plt.plot(data4[:,2],data4[:,3], '.-', color="purple")
plt.plot(data5[:,2],data5[:,3], '.-', color="indigo")
plt.xlabel("Desplazamiento. (mm)")
plt.ylabel("Fuerza. (kN)")
plt.title("jfvsfsf")
plt.legend()
plt.show()
```
